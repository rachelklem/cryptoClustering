```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# convert data type to int64
df_market_data['price_change_percentage_24h'] = df_market_data['price_change_percentage_24h'].astype(np.int64)
df_market_data['price_change_percentage_7d'] = df_market_data['price_change_percentage_7d'].astype(np.int64)
df_market_data['price_change_percentage_14d'] = df_market_data['price_change_percentage_14d'].astype(np.int64)
df_market_data['price_change_percentage_30d'] = df_market_data['price_change_percentage_30d'].astype(np.int64)
df_market_data['price_change_percentage_60d'] = df_market_data['price_change_percentage_60d'].astype(np.int64)
df_market_data['price_change_percentage_200d'] = df_market_data['price_change_percentage_200d'].astype(np.int64)
df_market_data['price_change_percentage_1y'] = df_market_data['price_change_percentage_1y'].astype(np.int64)

# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(market_data_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'])

# Copy the crypto names from the original data
coin_id = ["bitcoin", "ethereum", "tether", "ripple", "bitcoin-cash", "binancecoin", "chainlink", "cardano", "litecoin", "bitcoin-cash-sv", "crypto-com-chain", \
           "usd-coin", "eos", "monero", "tron", "tezos", "okb", "stellar", "cosmos", "cdai", "neo", "wrapped-bitcoin", "leo-token", "huobi-token", \
           "nem", "binance-usd", "iota", "vechain", "zcash", "theta-token", "dash", "ethereum-classic", "ethlend", "maker", "havven", "omisego", \
           "celsius-degree-token", "ontology", "ftx-token", "true-usd", "digibyte"]
coin_id_df = pd.DataFrame(coin_id)

# concatenate the columns?
scaled_df = pd.concat([coin_id_df, scaled_df], axis=1)

# Set the coin_id column as index
scaled_df.set_index(0, inplace=True)
scaled_df = scaled_df.rename_axis('coin_id')

# Display sample data
print(scaled_df)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = [k for k in range(1,11)]
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(scaled_df)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow = {'k': k_values, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
k_means = elbow_df.hvplot.line(x='k', y='inertia', title='Elbow Curve for K-means Clustering', xlabel='Number of Clusters (k)', ylabel='Inertia')
k_means
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model.predict(scaled_df)

# copy the preprocessed data
predictions_df = scaled_df.copy()

# somethin or another
predictions_df['predicted_clusters'] = k_4

# Print the resulting array of cluster values.
k_4
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

# Create the scatter plot

predictions_plot = predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="predicted_clusters",
    hover_cols=["coin_id"]
    )

# Display the scatter plot
predictions_plot

```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_data = pca.fit_transform(scaled_df)
# View the first five rows of the DataFrame. 
pca_data = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2', 'PCA3'])
pca_data.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is 0.38092 + 0.34765 + 0.17681, or 0.90538 (91%). 

```python
# Create a new DataFrame with the PCA data.
pca_data = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2', 'PCA3'])

# Copy the crypto names from the original data
coin_id = ["bitcoin", "ethereum", "tether", "ripple", "bitcoin-cash", "binancecoin", "chainlink", "cardano", "litecoin", "bitcoin-cash-sv", "crypto-com-chain", \
           "usd-coin", "eos", "monero", "tron", "tezos", "okb", "stellar", "cosmos", "cdai", "neo", "wrapped-bitcoin", "leo-token", "huobi-token", \
           "nem", "binance-usd", "iota", "vechain", "zcash", "theta-token", "dash", "ethereum-classic", "ethlend", "maker", "havven", "omisego", \
           "celsius-degree-token", "ontology", "ftx-token", "true-usd", "digibyte"]
coin_id_df = pd.DataFrame(coin_id)

# concatenate the columns?
pca_data = pd.concat([coin_id_df, pca_data], axis=1)

# Set the coinid column as index
pca_data.set_index(0, inplace=True)
pca_data = pca_data.rename_axis('coin_id')

# Display sample data
pca_data.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
pca_k_means = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in pca_k_means:
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    k_model.fit(pca_data)
    pca_inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow = {"pca_k_means": pca_k_means, "pca_inertia": pca_inertia}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow)

# check dataframe
pca_elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow = pca_elbow_df.hvplot.line(
    x="pca_k_means",
    y="pca_inertia",
    title="PCA Elbow Curve",
    color='blue',
    xticks=pca_k_means
)
pca_elbow
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
pca_model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
pca_model.fit(pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_predictions = pca_model.predict(pca_data)

# Print the resulting array of cluster values.
print(pca_predictions)
```

```python
# Create a copy of the DataFrame with the PCA data
pca_predictions_df = pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_predictions_df['pca_predictions'] = pca_predictions

# Display sample data
pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

pca_predictions_plot = pca_predictions_df.hvplot.scatter(
    x='PCA1', 
    y='PCA2',
    title="PCA Clustering",
    by='pca_predictions',
    hover_cols='coin_id'
)
pca_predictions_plot
```
